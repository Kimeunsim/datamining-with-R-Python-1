Define Boolean Conversion function
Using True or False


```python
def convert_boolean_columns(df):
  for col in df.columns:
    # Check if the column is of object type before attempting string operations
    if df[col].dtype == 'object':
        non_null = df[col].dropna().astype(str).str.lower()
        # t/f
        if non_null.isin(['t','f']).all():
            df[col] = df[col].astype(str).str.lower().map({'t': True, 'f': False})
        # True or False
        elif non_null.isin(['true','false']).all():
            df[col] = non_null.map({'true': True, 'false': False})
        # 0 or 1 (as strings)
        elif non_null.isin(['0','1']).all():
            df[col] = df[col].astype(int).astype(bool)
  return df
```

```python
#gsheet file url --> convert

def gsheet_to_df(gsheet_path):
    with open(gsheet_path, 'r') as f:
        for line in f:
            if "https://docs.google.com" in line:
                sheet_url = line.strip().replace('"', '')
                break
    csv_url = sheet_url.replace("/edit?usp=drive_web", "/export?format=csv")
    return pd.read_csv(csv_url)

```

```python
#import the data
#add both type of data csv and gsheet
def load_file(path):
    # Convert Path object to string before checking extension
    path_str = str(path)
    if path_str.endswith('.csv'):
        df = pd.read_csv(path)
    elif path_str.endswith('.gsheet'):
        df = gsheet_to_df(path)
    else:
        raise ValueError("error to load")
    df = convert_boolean_columns(df)
    return df
```

```python
#schema definition

schema="""
DROP TABLE IF EXISTS host;
DROP TABLE IF EXISTS listing;
DROP TABLE IF EXISTS amenities;
DROP TABLE IF EXISTS review;
DROP TABLE IF EXISTS calendar;

CREATE TABLE hosts(
  host_id INTEGER PRIMARY KEY, -- set host_id as primary key
  host_name TEXT,
  host_primary_verification TEXT
);

CREATE TABLE listings (
  id INTEGER PRIMARY KEY,
  host_id INTEGER,
  name TEXT,
  neighbourhood TEXT,
  room_type TEXT,
  price REAL,
  FOREIGN KEY(host_id) REFERENCES hosts(host_id)

);

CREATE TABLE amenities(
  amaenity_id INTEGER PRIMARY KEY AUTOINCREMENT,
  name TEXT UNIQUE
);

CREATE TABLE listing_amenities(
  listing_id INTEGER,
  amenity_id INTEGER,
  PRIMARY KEY (listing_id,amenity_id),
  FOREIGN KEY(listing_id) REFERENCES listings(id),
  FOREIGN KEY(amenity_id) REFERENCES amenities(amenity_id)

);

CREATE TABLE reviews(
  id INTEGER PRIMARY KEY AUTOINCREMENT,
  listing_id INTEGER,
  reviewer_name TEXT,
  comments TEXT,
  review_year INTEGER,
  FOREIGN KEY(listing_id) REFERENCES listings(id)
);

CREATE TABLE calendar(
  listing_id INTEGER,
  date TEXT,
  available INTEGER,
  price REAL,
  PRIMARY KEY (listing_id, date),
  FOREIGN KEY(listing_id) REFERENCES listings(id)
);
"""
```
